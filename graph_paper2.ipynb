```python
import matplotlib.pyplot as plt
import numpy as np

# Define graph paper dimensions in inches and DPI
width_inches = 8.5
height_inches = 11
dpi = 300  # High resolution for printing
width_px = int(width_inches * dpi)
height_px = int(height_inches * dpi)

# Define grid spacing and line thickness
# For 1/4 inch grid (standard graph paper)
grid_spacing_inches = 0.25
grid_spacing_px = int(grid_spacing_inches * dpi)  # 75 pixels between grid lines

# Thin lines for grid
line_thickness_px = 1  # Just 1 pixel thick

# Create a blank white canvas
graph_paper = np.ones((height_px, width_px, 3), dtype=np.uint8) * 255

# Calculate offsets to center the grid
x_offset = (width_px % grid_spacing_px) // 2
y_offset = (height_px % grid_spacing_px) // 2

# Draw the grid lines (centered)
for x in range(x_offset, width_px, grid_spacing_px):
    graph_paper[:, x:x+line_thickness_px] = 0  # Vertical lines

for y in range(y_offset, height_px, grid_spacing_px):
    graph_paper[y:y+line_thickness_px, :] = 0  # Horizontal lines

# Save the image as a PNG
file_path = "graph_paper.png"
plt.imsave(file_path, graph_paper)

print(f"Graph paper created with dimensions: {width_px}x{height_px} pixels")
print(f"Grid spacing: {grid_spacing_px} pixels ({grid_spacing_inches} inches)")
print(f"Line thickness: {line_thickness_px} pixel")
print(f"Grid centered with offsets: x={x_offset}, y={y_offset} pixels")
```
