```python
import matplotlib.pyplot as plt
import numpy as np

# Define graph paper dimensions in inches and DPI
width_inches = 8.5
height_inches = 11
dpi = 300  # High resolution for printing
width_px = int(width_inches * dpi)
height_px = int(height_inches * dpi)

width_px, height_px
```

```python
# Define grid spacing and line thickness in inches converted to pixels
grid_spacing_px = 56
line_thickness_px = 2

grid_spacing_px, line_thickness_px
```

```python
# Create a blank white canvas
graph_paper = np.ones((height_px, width_px, 3), dtype=np.uint8) * 255

# Draw the grid lines
for x in range(0, width_px, grid_spacing_px):
    graph_paper[:, x:x+line_thickness_px] = 0  # Vertical lines

for y in range(0, height_px, grid_spacing_px):
    graph_paper[y:y+line_thickness_px, :] = 0  # Horizontal lines

# Save the image as a PNG
file_path = "graph_paper.png"
plt.imsave(file_path, graph_paper, cmap="gray")

file_path
```
