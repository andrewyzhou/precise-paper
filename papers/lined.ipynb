```python
# Lined paper — three densities (regular / extra / super).
# Horizontal baselines on the canonical lattice. We SKIP line 0 (the
# topmost lattice line) because it only leaves 120 px of writing room
# above it — too tight for handwriting. First visible line is at j=1,
# giving a full SPACING of writing space between the header and the
# first baseline (same size as every subsequent row).
# Output: ../output/lined/{lined,extra-lined,super-lined}.png
from PIL import Image
from IPython.display import display
from common import (
    blank_page, draw_header, hline,
    grid_counts, lattice_y,
    output_path, variant_name, DENSITIES,
)

MODE = "lined"
```

```python
def render_lined(spacing):
    """Draw baselines at every lattice row EXCEPT j=0."""
    img = blank_page()
    _, n_y = grid_counts(spacing)
    for j in range(1, n_y):
        hline(img, lattice_y(j, spacing))
    draw_header(img)
    return img, n_y - 1
```

```python
for density, spacing in DENSITIES.items():
    img, n_lines = render_lined(spacing)
    path = output_path(MODE, variant_name(MODE, density))
    img.save(path, "PNG")
    print(f"{density:8s} spacing={spacing:3d}px  {n_lines:3d} lines  →  {path}")
    display(img.resize((850, 1100), Image.LANCZOS))
```
