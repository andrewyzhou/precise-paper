```python
# Lined paper — three densities (regular / extra / super).
# Horizontal lines on the canonical lattice. Same usable rectangle, header,
# and 4-side 120 px margins as every other mode. See common.py.
# Output: ../output/lined/{lined,extra-lined,super-lined}.png
from PIL import Image
from IPython.display import display
from common import (
    blank_page, draw_header, hline,
    grid_counts, lattice_y,
    output_path, variant_name, DENSITIES,
)

MODE = "lined"
```

```python
def render_lined(spacing):
    """Draw n_y horizontal lines at every lattice row."""
    img = blank_page()
    _, n_y = grid_counts(spacing)
    for j in range(n_y):
        hline(img, lattice_y(j, spacing))
    draw_header(img)
    return img, n_y
```

```python
for density, spacing in DENSITIES.items():
    img, n_y = render_lined(spacing)
    path = output_path(MODE, variant_name(MODE, density))
    img.save(path, "PNG")
    print(f"{density:8s} spacing={spacing:3d}px  {n_y:3d} lines  →  {path}")
    display(img.resize((850, 1100), Image.LANCZOS))
```
